```python
# Find Albums by singer
def research_Album(h):
    arr = np.array(Singer)
    #print(len(Singer))
    filter_arr = []
    for i in range(len(Singer)):
        if Singer[int(i)] == h:
            if Album[i]=='':
                continue
            h_Album = Album[i]
            # print(h_Album)
            filter_arr.append(h_Album)
            # print(type(h_Album))

        mode_Album = Counter(filter_arr).most_common(100)
            # print(mode_Album)
            # print(h_Album)
    return mode_Album
```
